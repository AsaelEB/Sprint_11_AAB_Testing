## PREPARAR LOS DATOS <a id='part_2'></a>

Antes de comenzar con la investigación requerida del proyecto, se realizará el análisis básico de los datos para comprobar que son correctos y no afecten en las investigaciones posteriores. En este preprocesamiento se realizan ligeros cambios al dataframe original, los cuáles nos ayudarán a manejar mejor los datos y poder realizar los análisis posteriores con mayor facilidad.

En caso de verse necesario se realizará un cambio en el nombre de las columnas. Se realizará una comprobación de los tipos de datos y valores ausentes. También se solicita que se agregue una columna de fecha y hora, así como una columna separada para las fechas.

Para que se tenga una idea clara de los datos con los que se trabaja en un inicio, se describe el archivo en el que se registraron los resultados obtenidos de las pruebas A/A/B; cada fila en el registro es la acción de un usuario o un 'evento'. El archivo se descargó con la ruta: `/datasets/logs_exp_us.csv` y las columnas que contiene son las siguientes:

* EventName: Nombre del evento.
* DeviceIDHash: Identificador de usuario único.
* EventTimestamp: Hora del evento.
* ExpId: Número de experimento: 246 y 247 son los grupos de control (A/A), 248 es el grupo de prueba(B).


```python
# Importar librerias
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats as sts
from statsmodels.stats.weightstats import ztest as ztest
```

```python
# Cargar el archivo y renombrar columnas
df = pd.read_csv("datasets/logs_exp_us.csv", sep='\t' )
df.rename(columns={'EventName':'event_name', 'DeviceIDHash':'device_id', 'EventTimestamp':'event_ts', 'ExpId':'group'}, inplace=True)

# Timestamp a fecha
df['event_ts'] = pd.to_datetime(df['event_ts'], unit='s', origin='unix')
```

```python
# Observar los tipos de datos, elementos nulos y duplicados
print(df.describe())
print()
print(df.info())
print()
print("Duplicados: ", df.duplicated().sum())
```

```python
# Eliminar elementos duplicados
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(df.info())
```

```python
# Agregar una columna de fecha y hora y una columna separada para las fechas
df['event_date'] = df['event_ts'].dt.date
df['event_time'] = df['event_ts'].dt.time
display(df.head())
```

**CONCLUSIONES**

Se han cambiado los nombres de las columnas a valores más explicativos y sencillos de trabajar. Se observaron 413 filas duplicadas que posteriormente se eliminaron, en principio no se observaron valores nulos. La columna de timestamp fue cambiada a una columna de tipo fecha para que pudiera ser leída apropiadamente y, como se solicitó, se añadieron dos columnas extras, una que contenga sólamente la fecha del evento y otra con la hora del evento.

### Cantidad de usuarios para cada evento <a id='users_quant_and_proportions'></a>
Ahora que conocemos los eventos que hay y el número de veces que se repiten, se realizará la busqueda de la cantidad de usuarios que realizaron cada una de estas acciones. Para anlizar los datos se ordenan los eventos por el número de usuarios. De igual manera se pide calcular la proporción de usuarios que realizaron la acción al menos una vez.

```python
# Cantidad total de usuarios en los datos
print("Usuarios únicos registrados en los datos:", new_df['device_id'].nunique())

# Usuarios únicos para cada evento
users_proportion = new_df.groupby('event_name', as_index=False).agg({'device_id':pd.Series.nunique}).sort_values(by='device_id', ascending=False).reset_index(drop=True)

# Proporción de los usuarios que han realizado los eventos
users_proportion['proportion'] = users_proportion['device_id'] / new_df['device_id'].nunique()
display(users_proportion)
```

**CONCLUSIONES**

Se observa que la cantidad de usuarios hace que los eventos tengan el mismo orden desde el evento `Main Screen Appear` hasta el evento `Tutorial` con menor número de usuarios que realizaron este evento.

En cuanto a las proporciones la cantidad de usuarios totales es de 7534, de acuerdo a los resultados, sólamente 7419 realizaron el evento `Main Screen Appear`, dado esto, puede significar que es el primer evento que se les muestra a los usuarios. Menor cantidad de usuarios se registraron en los eventos posteriores.

### Proporción de usuarios que pasan de etapa <a id='stage_users'></a>
La tarea de esta sección consiste en utilizar el embudo de eventos para encontrar la proporción de usuarios que pasan de una etapa a la siguiente y observar en qué etapa se pierden más usuarios. Por ejemplo, para la secuencia de eventos A → B → C, calcular la proporción de usuarios en la etapa B en relación a la cantidad de usuarios en la etapa A y la proporción de usuarios en la etapa C en relación a la cantidad en la etapa B.

Para realizar esta operación, haré uso de una tabla dinámica que me ayudará a observar a la hora minima de realización de los eventos para cada usuario, de esta manera, al contabilizar las etapas se obtendrá el número de usuarios que pasan de una etapa a la otra siguiendo la ruta establecida.

```python
# Tabla dinamica de usuarios y eventos minimos
users = new_df.pivot_table(index='device_id',  columns='event_name',  values='event_ts', aggfunc='min')

# Secuencia de eventos que se debieron seguir, se usa el Tutorial como último evento
step_1 = ~users['MainScreenAppear'].isna() 
step_2 = step_1 & (users['OffersScreenAppear'] > users['MainScreenAppear']) 
step_3 = step_2 & (users['CartScreenAppear'] > users['OffersScreenAppear']) 
step_4 = step_3 & (users['PaymentScreenSuccessful'] > users['CartScreenAppear']) 
step_5 = step_4 & (users['Tutorial'] > users['PaymentScreenSuccessful']) 

n_mains = users[step_1].shape[0] 
n_offers = users[step_2].shape[0] 
n_carts = users[step_3].shape[0] 
n_payments = users[step_4].shape[0] 
n_tutorials = users[step_5].shape[0] 

users_proportion['users_passed'] = [n_mains, n_offers, n_carts, n_payments, n_tutorials]
users_proportion['users_passed_proportion'] = users_proportion['users_passed'] / users_proportion['device_id']

# Agregar el porcentaje de usuarios que pasan a la siguiente etapa
percent = []
for i in users_proportion.index:
    if i > 0:
        percent.append((users_proportion['users_passed'][i] * 100) / users_proportion['users_passed'][i-1])
    else:
        percent.append((users_proportion['users_passed'][i] * 100) / users_proportion['users_passed'][i])
users_proportion['users_passed_percent'] = percent
        

display(users_proportion)
```

**CONCLUSIONES**

Se observa que solamente un usuario concluyó de la manera en la que se sugirió la ruta; dado que el TUTORIAL puede ser opcional se toma el evento `Payment Screen Successful` como el último evento de la ruta a seguir, lo habrán concluido un total de 454 usuarios, de los 7419 que iniciaron la ruta definida.

También se logra apreciar que la proporción se reduce drasticamente del segundo al tercer evento, esto quiere decir que muchos usuarios se quedaron en el evento `Offers Screen Appear` siguiendo la ruta establecida (3218 usuarios de diferencia).

Sin embargo, usando el porcentaje de usuarios que pasaron de etapa, se observa que entre el evento `Main Screen Appear` y `Offers Screen Appear` hay una mayor diferencia de usuarios. Siendo el evento MAIN SCREEN en el que se perdieron la mayoría de los usuarios.

### Diferencia entre los grupos A <a id='stats_difference'></a>
Tenemos dos grupos de control en el test A/A, donde comprobamos nuestros mecanismos y cálculos. La tarea es observar si existe una diferencia estadísticamente significativa entre las muestras de los grupos 246 y 247, esto es, observar la cantidad de usuarios que llegan a un evento, en este caso el evento final `Payment Screen Succesfull`.

He realizado una búsqueda de los usuarios que tienen el evento requerido y he usado el valor 1 para representar que son usuarios que llegaron a ese evento. El resto de usuarios se representan con valor 0 en la columna de conversión. Al final se usa un Series con 1 y 0 con el tamaño de usuarios únicos en el grupo para realizar las pruebas.

Usando una prueba z-test se podrá mostrar si es que existe una diferencia entre la conversión de los grupos de control. Plantenado las hipótesis:

Hipótesis Nula: No existe una diferencia significativa entre los grupos.

Hipótesis alternativa: Existe diferencia estadísticamente significativa entre los grupos.

```python
# Función para crear la conversion de un grupo
def conversion_sample(event_data, usuarios, event_name = 'PaymentScreenSuccessful'):
    # Usuarios que llegaron al evento
    converted_id = event_data[event_data['event_name'] == event_name]['device_id'].unique()
    # DataFrame de id y columna de conversion
    converted = pd.DataFrame(data = {'device_id' : converted_id, 'conversion': 1})
    # Agregar usuarios que no se convirtieron y llenar con 0 en la columna de conversión
    conversion = usuarios.merge(converted,on ='device_id' ,how = 'left')
    conversion.fillna(0, inplace = True)
    #print(conversion)
    return conversion

# Función para realizar la prueba Z-Test
def z_test(group_1, group_2, alpha=0.05):
    p_value = ztest(group_1, group_2)[1]
    print('p-value:', p_value)
    if p_value < alpha:
        print('Se rechaza la Hipótesis Nula')
    else:
        print('No se puede rechazar la Hipótesis Nula')
```

```python
# Usuarios únicos totales
total_users = new_df[['device_id','group']].drop_duplicates()
# DataFrame con todos usuarios y su conversión
conversiones = conversion_sample(new_df, total_users)
# Usuarios por grupos con conversión
payment_conversion_246 = conversiones[conversiones['group']==246]
payment_conversion_247 = conversiones[conversiones['group']==247]

display(payment_conversion_246)
display(payment_conversion_247)
```

```python
z_test(payment_conversion_246['conversion'], payment_conversion_247['conversion'])
```

**CONCLUSIONES**

Si la prueba realizada es correcta, entonces no se rechaza la hipótesis nula porque no existe una diferencia estadísticamente significativa entre estos dos grupos, usando el evento Payment Screen Successful como punto al que deben llegar los usuarios.

Después usaré la función para hacer una prueba de tipo Z-test para comprobar si existe una diferencia entre los grupos proponiendo las siguientes hipótesis:

Hipótesis Nula: No existe una diferencia significativa entre los grupos.

Hipótesis alternativa: Existe diferencia estadísticamente significativa entre los grupos.

Si la hipótesis nula no es rechazada, significaría que los grupos se han dividido correctamente, en caso de rechazar la hipótesis nula, se puede pensar que los grupos no fueron divididos correctamente.

```python
# Función para la proporción de usuarios convertidos y no convertidos
def proportion(group):
    grouped = group.groupby('conversion').count()
    grouped['proportion'] = grouped['device_id'] / grouped['device_id'].sum()
    return grouped

# Función para hacer los cálculos para cada Evento en 2 grupos
def all_events_difference(group_1, group_2, alpha=0.05):
    for event in users_proportion['event_name'].unique():
        # Buscar la conversión para cada evento
        conversiones = conversion_sample(new_df, total_users, event)
        # Usuarios por grupos con conversión
        conversion_a1 = conversiones[conversiones['group']==group_1]
        conversion_a2 = conversiones[conversiones['group']==group_2]
        print('\nCONVERSIÓN PARA EL EVENTO:', event)
        display(proportion(conversion_a1))
        display(proportion(conversion_a2))
        z_test(conversion_a1['conversion'], conversion_a2['conversion'], alpha)
```
